```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('nba_2017_players_stats_combined.csv')
```

```python
player_age = df.loc[:, 'AGE']

plt.boxplot(player_age)
plt.show()
```

The boxplot see in cell 3 above is of all the ages of NBA players in the 2016-17 season. The meadian/average age of a player is about 26 years old, and about 50% of players are between the ages of 24-28. However, there are a few outliers who are closer to age 40.

```python
player_rank = df.loc[:, 'Rk']

plt.scatter(player_age, player_rank, color='purple')
plt.ylabel('Rank')
plt.xlabel('Age')
plt.title('Rank by Age')

plt.show()
```

This scatter plot shows a heavy concentration of top 100 ranked players being between the ages of 24 and 28. That matches the average age range of players in the league, and could lead to saying players between the ages of 24 and 28 are at the top of their game. However there are other factors at steak, which is why some older players (30+) are still ranked in the top 100.

```python
games_played = df.loc[:30, 'GP']
games_won = df.loc[:30, 'W']

plt.plot(games_played)
plt.plot(games_won)
plt.ylabel('Games played')
plt.title('Games Played vs Games Won by Player')
plt.legend()
plt.show()
```

Most players play a consistent amount of games so long as they're not injury prone or dealing with other factors that would have them out of a game. There's a dramatic drop at the end of the season because not every team makes the playoffs. Games won shows that players and teams go through winning and losing streaks, which follows the dramatic ups and downs seen in the graph.

```python
plt.figure(figsize=(15, 5))

fg_percent = df.loc[:, 'FG%']

plt.subplot(1, 3, 1)
plt.hist(player_age, color='blue')
plt.ylabel('Frequency')
plt.title('Player Age')

plt.subplot(1, 3, 2)
plt.hist(fg_percent, color='red')
plt.ylabel('Frequency')
plt.title('FG Percentages')

plt.subplot(1, 3, 3)
plt.scatter(player_age, fg_percent, color='pink')
plt.ylabel('FG%')
plt.title('FG% by Age')

plt.show()
```

These subplots show a more easily read breakdown of ages, average field goal percentages, and field goal percentage by age. The first histogram reiterates that most players are about 24, the second histogram shows that most players shoot at just over 40% efficiency, which could lead people to think that most players who are around 24 years, shoot at just over 40%. The scatter plot proves that regardless of age, most players are shooting around 40%.
