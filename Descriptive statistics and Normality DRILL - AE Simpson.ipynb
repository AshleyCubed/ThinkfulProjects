```python
# Generate a binomial distribution with n=20 and p=0.5.
binom = np.random.binomial(20, 0.5, 100)

#Plot a histogram.
plt.hist(binom)

# Print the histogram
plt.show()
```

```python
#Plot a histogram for binom.
plt.hist(binom, bins=10, color='c')

# Add a vertical line at the mean.
plt.axvline(binom.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(binom.mean() + binom.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(binom.mean()-binom.std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()
```

*Notes: The mean is also the mode, with the most occurances happening at 10. The standard deviation is about 2.5 (by just looking at the histogram).

```python
# Generate a f-distribution with degrees of freedom between 15 and 45
dfnum = 15. # between group degrees of freedom
dfden = 45. # within groups degrees of freedom
f_dist = np.random.f(dfnum, dfden, 100)

# Plot a histogram.
plt.hist(f_dist)

# Print the histogram.
plt.show()
```

```python
#Plot a histogram for f-dsitribution.
plt.hist(f_dist, bins=10, color='r')

# Add a vertical line at the mean.
plt.axvline(f_dist.mean(), color='orange', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(f_dist.mean() + f_dist.std(), color='orange', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(f_dist.mean()-f_dist.std(), color='orange', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()
```

<b>Evaluate whether the descriptive statistics provided useful information about the variable. Can you identify any common characteristics of the distributions that could be usefully described using the mean and/or standard deviation, versus the ones that could not?</b><br/>
Chisquare, f, and logseries distributions are all positively skewed distributions. So the mean will be where most of the data is clustered, but only 1 standard deviation above the mean will show any data, but it won't be relative to the data found at a standard deviation below the mean. <br/>
On the other hand, Poisson, binomial, and gamma distributions follow fairly normal distributions with minimal skew and are easily able to show central tendency.

<b>Additionally: </b><br/>

Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.<br/>
Add them together to create a third variable.<br/>
Graph the third variable using a histogram.<br/>
Compute the mean and standard deviation and plot them as vertical lines on the histogram.<br/>
Evaluate the descriptive statistics against the data.<br/>

```python
var1 = np.random.normal(5,.5, 100)
var2 = np.random.normal(10,1, 100)
var3 = var1+var2

#Plot a histogram for variable 3.
plt.hist(var3, color='c')

# Add a vertical line at the mean.
plt.axvline(var3.mean(), color='black', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(var3.mean() + var3.std(), color='y', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(var3.mean()-var3.std(), color='y', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()
```

Adding the two normal distributions still leaves the mean where the data clusters the most (mode), and there is a slight positive skew.
